```python
df[df.price_per_sqft>df.price_per_sqft.quantile(.80)].head(10)
```

```python
outlier_no=df[df.price_per_sqft<df.price_per_sqft.quantile(.80)]
```

```python
#histplot to check the normality of the column price per sqft column.

sns.histplot(data=outlier_no,x=outlier_no['price_per_sqft'],kde=True)
plt.show()
```

```python
IQR=Q3-Q1
IQR
```

```python
LOWER=Q1-1.5*IQR
UPPER=Q3-1.5*IQR
#OUTLIERS:

df[(df.price_per_sqft<LOWER)|(df.price_per_sqft>UPPER)].head()
```

```python
without_outliers=df[(df.price_per_sqft>LOWER)&(df.price_per_sqft<UPPER)]
without_outliers.head()
```

```python
sns.histplot(data=without_outliers,x=without_outliers['price_per_sqft'],kde=True)
plt.show()
```

```python
#Z-SCORE METHOD

M=df.price_per_sqft.mean()
S=df.price_per_sqft.std()
lower=M-3*S
upper=M+3*S
```

```python
df['ZScore']=(df.price_per_sqft-M)/S
```

```python
#outliers
df[(df.ZScore<-3)|(df.ZScore>3)]
```

```python
df1=df[(df.price_per_sqft>lower)&(df.price_per_sqft<upper)]
```

```python
#to checck it is normal distribution
sns.histplot(data=df1,x=df1['price_per_sqft'],kde=True)
plt.show()
```

```python
#box plot for numerical columns

plt.figure(figsize=(10,5))
sns.boxplot(data=outlier_no)
plt.xticks(rotation=45)
plt.title('box plot for numerical columns')
plt.show()
```

```python
#correlation b/w numerical columns
correlation_columns=outlier_no.drop(['location', 'size'],axis=1)
matrix=correlation_columns.corr()
sns.heatmap(data=matrix,annot=True,cmap='spring')
plt.show()
```

```python
#Scatter plot between the variables
sns.pairplot(data=correlation_columns,hue='bhk')
plt.title('pairplot')
plt.show()
```
